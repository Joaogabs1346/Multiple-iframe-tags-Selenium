### Open the web Site:

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
driver = webdriver.Chrome()
driver.get('https://pbdatatrader.com.br/jogosdodia')

```

### Waiting for the page to load and opening items

```python
#Wait one element to open
while len(driver.find_elements(by=By.TAG_NAME,value='iframe')) == 0: 
    time.sleep(1)
#Open the first iframe
first_iframe= driver.find_element(by=By.TAG_NAME,value='iframe')
driver.switch_to.frame(first_iframe)
#Open the second Iframe
first_iframe= driver.find_element(by=By.TAG_NAME,value='iframe')
driver.switch_to.frame(first_iframe)


```

### Open the data table

```python

table = driver.find_element(by=By.CLASS_NAME, value='innerContainer').text

```

### Formatting the table

```python
import pandas as pd

lista_texto= table.split("\n")
colunas = lista_texto[:11]
valores = lista_texto[11:]

dic= {}
for coluna in colunas:
    dic[coluna] = []
    for i in range(20):
        dic[coluna].append(valores[i])
    valores = valores[20:] 
    

    
table_df = pd.DataFrame.from_dict(dic)
display(table_df)
```

### Implementing the scroll


```python
#How find the scroll:
'''
var elements = document.getElementsByTagName("div");
for(var i = 0; i < elements.length; i++) { if(elements[i].scrollHeight> document.getElementsByTagName("body")[0].scrollHeight){
    console.log(elements[i]);
    console.log(elements[i].scrollHeight);
}}
'''
for i in range(400):
    posicao_scroll=250 * i
    driver.execute_script(f'document.getElementsByClassName("bodyCells")[0].scroll(0,{posicao_scroll})')
    #read the page table   
    table = driver.find_element(by=By.CLASS_NAME, value='innerContainer').text
    #Separating values and columns
    lista_texto= table.split("\n")
    colunas = lista_texto[:11]
    valores = lista_texto[11:]
    #organize table
    dic= {}
    for coluna in colunas:
        dic[coluna] = []
        for i in range(20):
            dic[coluna].append(valores[i])
        valores = valores[20:] 

      #creating the table  
    full_table_df = pd.DataFrame.from_dict(dic)
    table_df= table_df.append(full_table_df,ignore_index=True)


    

  
```

### Removing Duplicates and Creating a Spreadsheet in Excel

```python
#Remover duplicatas
table_df = table_df.drop_duplicates()
#Exibir a tabela full e transformar para  o excel
display(table_df)

```
